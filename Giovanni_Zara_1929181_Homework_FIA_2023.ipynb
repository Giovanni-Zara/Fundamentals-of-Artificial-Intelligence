### Metrics

#### Regression
To evaluate how good we are doing on the **regression** task we will use the Root Mean Squared Error (RMSE). RMSE is given by

$$
\sqrt{\frac{1}{n}\sum\limits_{i=1}^{n}{\Big(d_i -f_i\Big)^2}}
$$


where:

* $n$ is the number of test samples
* $d_i$ is the ground truth value of the i-th sample
* $f_i$ is the predicted value of the i-th sample


#### Classification
To evaluate how good we are doing on the **classification** task we will use the accuracy metrics. Accuracy is given by

$$
\frac{TP+TN}{TP+TN+FP+FN}
$$

where:

* TP is the number of *correctly* classified positive samples
* TN is the number of *correctly* classified negative samples
* FP is the number of *incorrectly* classified positive samples
* FN is the number of *incorrectly* classified negative samples

#Procedo di seguito con l'implementazione del mio Decision Tree.

* La classe Node rappresenta un nodo dell'albero, mentre la classe DecisionTree contiene i metodi per costruire e utilizzare l'albero.

* Nel metodo fit, viene costruito l'albero utilizzando il metodo _build_tree, che utilizza la ricorsione per generare i nodi dell'albero in base alle migliori divisioni trovate per le feature.

* Il metodo _best_split trova la migliore feature e soglia di separazione utilizzando il criterio di guadagno d'informazione.

* Il metodo _create_split suddivide il dataset in base alla soglia di separazione trovata.

* Il metodo predict utilizza l'albero per fare previsioni su nuovi dati utilizzando la funzione _traverse_tree, che attraversa l'albero in base alle soglie di separazione per trovare il valore di output corrispondente.

```python
import math
class Node:
  def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
    self.feature = feature
    self.threshold = threshold
    self.left = left
    self.right = right
    self.value = value

  def is_leaf(self):
    return self.value is not None

class DecisionTree:
  def __init__(self, max_depth=100, min_samples_split=2):
    self.max_depth = max_depth
    self.min_samples_split = min_samples_split
    self.root = None
    self.used_features = set()


  def _is_finished(self, depth):
    if (depth >= self.max_depth
      or self.n_class_labels == 1
      or self.n_samples < self.min_samples_split):  #se il n dei samples è minore di quello dei samples splittati ti devi fermare
      return True
    return False


  def fit(self, X, y):
      self.root = self._build_tree(X, y)

  def _entropy(self, y):
      proportions = np.bincount(y) / len(y)
      entropy = -np.sum([p * np.log2(p) for p in proportions if p > 0])
      return entropy

  def _create_split(self, X, thresh):
      left_idx = np.argwhere(X <= thresh).flatten()
      right_idx = np.argwhere(X > thresh).flatten()
      return left_idx, right_idx

  def _information_gain(self, X, y, thresh):
      parent_loss = self._entropy(y)
      left_idx, right_idx = self._create_split(X, thresh)
      n, n_left, n_right = len(y), len(left_idx), len(right_idx)

      if n_left == 0 or n_right == 0:
          return 0

      child_loss = (n_left / n) * self._entropy(y[left_idx]) + (n_right / n) * self._entropy(y[right_idx])
      return parent_loss - child_loss

  def _best_split(self, X, y, features):
      split = {'score':- 1, 'feat': None, 'thresh': None}

      for feat in features:
          X_feat = X[:, feat]
          thresholds = np.unique(X_feat)
          for thresh in thresholds:
              score = self._information_gain(X_feat, y, thresh)

              if score > split['score']:
                  split['score'] = score
                  split['feat'] = feat
                  split['thresh'] = thresh

      return split['feat'], split['thresh']

  def _traverse_tree(self, x, node):  #funzione che attraversa l'albero
        if node.is_leaf():
            return node.value

        if x[node.feature] <= node.threshold:
            return self._traverse_tree(x, node.left)
        return self._traverse_tree(x, node.right)

  def predict(self, X):
      predictions = [self._traverse_tree(x, self.root) for x in X]
      return np.array(predictions)

  def _build_tree(self, X, y, depth=0, i=0):
      self.n_samples, self.n_features = X.shape
      self.n_class_labels = len(np.unique(y))

      # criterio per capire quando fermarsi, poi dopo quando faccio la grow ricorsiva aggiungo 1 alla depth ogni iterazione. Quindi si ferma anche se la depth raaggiunge quella massima da me impostata
      if self._is_finished(depth):
          most_common_Label = np.argmax(np.bincount(y))
          return Node(value=most_common_Label)

      #prendere il best split
      '''rnd_feats = np.random.choice(self.n_features, self.n_features, replace=False) #qui ci devo mettere le mie features non delle random!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
      best_feat, best_thresh = self._best_split(X, y, rnd_feats)'''
      best_feat, best_thresh = self._best_split(X, y, range(self.n_features))



      # ora una grow dei figli ricorsiva
      left_idx, right_idx = self._create_split(X[:, best_feat], best_thresh)
      left_child = self._build_tree(X[left_idx, :], y[left_idx], depth + 1)
      right_child = self._build_tree(X[right_idx, :], y[right_idx], depth + 1)
      return Node(best_feat, best_thresh, left_child, right_child)
```

#implementazione
Il codice rappresenta un'implementazione del modello di regressione lineare utilizzando l'algoritmo di gradient descend per l'addestramento.

* Nel costruttore della classe LinearRegression, vengono inizializzati i parametri principali per l'addestramento del modello: learning_rate (il tasso di apprendimento), num_iterations (il numero massimo di iterazioni) e tol (la tolleranza per il criterio di convergenza).

* Nel metodo fit, viene effettuato l'addestramento del modello nel seguente modo:
   - La matrice delle features X viene estesa con una colonna di 1s (cioè di tutti 1) per rappresentare il termine di intercetta.
   - I pesi del modello self.weights vengono inizializzati con valore zero.
   - Viene iterato per un numero massimo di iterazioni specificato. Durante ogni iterazione:
     - Viene calcolata la previsione hypothesis moltiplicando la matrice delle features X per i pesi self.weights.
     - Viene calcolato l'errore tra la previsione e le label y.
     - Viene calcolato il gradiente moltiplicando la trasposta della matrice delle features X.T per l'errore e dividendo per il numero di campioni m.
     - I pesi self.weights vengono aggiornati sottraendo il prodotto tra il tasso di apprendimento learning_rate e il gradiente.
     - Viene calcolata la perdita (MSE) come la somma dei quadrati degli errori divisa per il doppio del numero di campioni m.
     - Viene verificata la convergenza confrontando la differenza tra la perdita corrente e la perdita precedente con la tolleranza tol. Se la differenza è inferiore alla tolleranza, l'addestramento viene interrotto.
* Il metodo predict viene utilizzato per effettuare le previsioni utilizzando il modello addestrato. La matrice delle features X viene estesa con una colonna di 1s e quindi viene effettuato il prodotto scalare tra questa matrice e i pesi self.weights per ottenere le previsioni.


```python
class LinearRegression:

    def __init__(self, learning_rate, num_iterations, tol):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.tol = tol

    def fit(self, X, y):
        m, n = X.shape  #shape restituisce righe,colonne della matrice
        X = np.concatenate((np.ones((m, 1)), X), axis=1) #np.ones crea una matrice di dimensioni (m, 1) contenente tutti 1. Questa matrice rappresenta una colonna di 1s che rappresenta il termine di intercetta del modello di regressione lineare.
        self.weights = np.zeros(n+1)                    #np.concatenate concatena la matrice di 1s alla matrice delle features X lungo l'asse 1 (cioè aggiunge le colonne insieme). Il risultato è una nuova matrice in cui la colonna di 1s viene aggiunta come prima colonna e le colonne di X seguono.
        loss_precedente = np.inf

        for i in range(self.num_iterations):
            ipotesi = np.dot(X, self.weights)
            errore = ipotesi - y
            gradient = np.dot(X.T, errore) / m
            self.weights -= self.learning_rate * gradient

            loss = np.sum(errore**2) / (2 * m)
            if abs(loss - loss_precedente) < self.tol:
                break
            loss_precedente = loss


    def predict(self, X):
        m = X.shape[0]
        X = np.concatenate((np.ones((m, 1)), X), axis=1)
        return np.dot(X, self.weights)


```

#Regressione lineare con regolarizzazione Lasso Regression (L1) (algoritmo di gradient descend).

Questo codice implementa la regressione Lasso, che è una tecnica di regressione lineare con regolarizzazione L1.

In questa implementazione:
* il costruttore della classe prende in input i parametri alpha, lr e num_iterations, che rappresentano rispettivamente il parametro di regolarizzazione L1, il tasso di apprendimento utilizzato nell'algoritmo di gradient descent e il numero di iterazioni che il metodo fit eseguirà durante l'apprendimento.

* Il metodo fit esegue l'addestramento del modello. Riceve in input la matrice delle feature X e il vettore delle label y. Inizia inizializzando i pesi self.weights come un array di zeri della stessa dimensione delle feature.
Successivamente, il metodo itera per num_iterations, eseguendo il seguente processo:

  - Calcola le previsioni y_pred moltiplicando la matrice delle feature X per i pesi self.weights.
  - Calcola l'errore tra le previsioni y_pred e le etichette vere y.
  - Calcola la penalità L1 moltiplicando il parametro di regolarizzazione alpha per il segno dei pesi self.weights utilizzando np.sign.
  - Calcola il gradiente come la somma di due componenti: la derivata dell'errore rispetto ai pesi e la penalità L1. Il termine (-1/m) * np.dot(X.T, error).-

  - Calcola la derivata dell'errore rispetto ai pesi usando il gradiente del mean squared error. Il termine l1_penalty rappresenta la penalità L1.
  - Aggiorna i pesi self.weights utilizzando l'algoritmo di gradient descent, moltiplicando il gradiente per il learning rate lr e sottraendo il risultato ai pesi correnti.
  Il metodo predict esegue la predizione utilizzando i pesi self.weights appresi durante la fase di addestramento. Moltiplica la matrice delle feature X per i pesi per ottenere le previsioni.

```python
class LassoRegression:

    def __init__(self, alpha, lr, num_iterations):  #alpha è parametro di regolarizzazione L1, lr = learning rate,
                                                                  #num_iterations serve a dire quante iterazione far fare al gradient descend
        self.alpha = alpha
        self.lr = lr
        self.num_iterations = num_iterations

    def fit(self, X, y):
        m, n = X.shape
        self.weights = np.zeros(n)  #creo un array di pesi inizializzato a zero
        for i in range(self.num_iterations):
            y_pred = np.dot(X, self.weights)
            error = y - y_pred
            l1_penalty = self.alpha * np.sign(self.weights) #np.sign mi returna -1 se weight<0, 0 se weight=0, 1 se weight>0
            grad = (-1/m) * np.dot(X.T, error) + l1_penalty
            self.weights -= self.lr * grad

    def predict(self, X):
        return np.dot(X, self.weights)  #np.dot calcola prodotto scalare tra due array multidimensionali, la moltiplicazione della matrice delle features X per i pesi appresi durante la fase di fitting del modello
```

#regressione lineare con regolarizzazione L2 (ridge regression)

L'algoritmo di Gradient Descent per la regressione lineare con regolarizzazione L2, anche noto come Ridge Regression, consiste nell'aggiungere un termine di regolarizzazione L2 alla funzione costo originale. La funzione costo della regressione lineare con regolarizzazione L2 è data da:

J(w) = (1/2m) * [ sum((h(x(i))-y(i))^2) + beta * sum(w(j)^2) ]

dove:

* h(x(i)) rappresenta la previsione della variabile dipendente (output) per il campione x(i) (l'ennesima riga del dataset), calcolata come prodotto scalare tra i coefficienti del modello w e la riga x(i)
* y(i) rappresenta il valore reale dell'output per il campione x(i)
* m rappresenta il numero di campioni (righe del dataset)
lambda rappresenta il coefficiente di regolarizzazione L2, che controlla la forza della regolarizzazione
* w(j) rappresenta il coefficiente j-esimo del modello, con j=1,2,...,n, dove n rappresenta il numero di feature (colonne del dataset)
La regolarizzazione L2 ha l'effetto di penalizzare i coefficienti del modello w con un termine quadratico, limitandone la magnitudo e prevenendo l'overfitting

Per implementare l'algoritmo di Gradient Descent con regolarizzazione L2, utilizzo una procedura simile a quella senza regolarizzazione. In particolare, aggiorno i pesi del modello in questo modo:

w(j) = w(j) - lr * [(1/m) * sum((h(x(i))-y(i))*x(i,j)) + (beta/m)*w(j)]

dove:

* lr rappresenta il learning rate, ovvero la grandezza del passo di aggiornamento
* x(i,j) rappresenta l'elemento nella riga i-esima e colonna j-esima del dataset

Nella seguente implementazione:
* __init__ è il costruttore della classe che inizializza gli attributi dell'istanza. beta rappresenta il parametro di regolarizzazione L2, lr è il learning rate utilizzato nell'algoritmo di gradient descend e num_iterations è il numero di iterazioni per l'algoritmo.
* fit(self, X, y) è il metodo che addestra il modello di regressione ridge utilizzando i set di dati X e y. Prima di iniziare l'addestramento, viene aggiunta una colonna di 1 al dataset X per rappresentare il termine di intercetta. Quindi, i pesi del modello vengono inizializzati a zero. Successivamente, viene eseguito l'algoritmo di gradient descent per un numero fissato di iterazioni. Durante ogni iterazione, vengono calcolate le predizioni y_pred moltiplicando la matrice delle features X per i pesi correnti. Viene calcolato l'errore tra le predizioni e i valori effettivi y. Viene applicata la regolarizzazione L2 per limitare l'overfitting, aggiungendo il termine di regolarizzazione ai gradienti calcolati. Infine, i pesi vengono aggiornati utilizzando il learning rate moltiplicato per il gradiente calcolato.
* predict(self, X) è il metodo che effettua le predizioni utilizzando il modello addestrato. Prima di fare le predizioni, viene aggiunta una colonna di 1 al dataset X sempre per rappresentare il termine di intercetta. Viene quindi calcolato il prodotto scalare tra la matrice delle features X e i pesi appresi durante la fase di fitting del modello. Il risultato è l'array delle predizioni.

```python
class L2Regression:

    def __init__(self, beta=1.0, lr=0.01, num_iterations=1000):
        self.lr = lr   #learning rate
        self.beta = beta  #parametro di regolarizzazione L2, è usato per controllare la complessità del modello. se valore più alto porta a modello più semplice, valore più basso porta a modello più complesso.
        self.num_iterations = num_iterations
        self.weights = None

    def fit(self, X, y):
        # Aggiunge una colonna di 1 al dataset (bias)
        X = np.c_[np.ones((X.shape[0], 1)), X]

        # metti l'array dei pesi del modello a zero
        self.weights = np.zeros(X.shape[1])

        # algoritmo di gradient descend eseguito numero di volte che imposto come iperparametro
        for i in range(self.num_iterations):
            y_pred = np.dot(X, self.weights)
            error = y_pred - y
            regularization = (self.beta / X.shape[0]) * self.weights
            regularization[0] = 0
            gradient = (1 / X.shape[0]) * np.dot(X.T, error) + regularization
            self.weights -= self.lr * gradient

    def predict(self, X):
        #aggiungo colonna di 1 per il termine di intercetta
        X = np.c_[np.ones((X.shape[0], 1)), X]
        #faccio prodotto scalare tra matrice delle features e i pesi
        return np.dot(X, self.weights)
```

#regressione lineare elastic net

L'elastic regression è una tecnica di regressione che combina la regolarizzazione L1 (Lasso) e L2 (Ridge) per ridurre l'overfitting e migliorare la generalizzazione del modello.

La classe ha quattro parametri iniziali: learning_rate, num_iterations, l1_penality (penalità L1) e l2_penality (penalità L2). Questi parametri controllano il processo di addestramento del modello.

Come sempre il metodo fit esegue l'addestramento del modello sfruttando l'algoritmo di gradient descent. Viene inizializzato un vettore di pesi (self.W) e un termine di bias (self.b). L'aggiornamento dei pesi viene eseguito iterativamente nel metodo update_weights utilizzando i gradienti calcolati. Il metodo predict viene utilizzato per effettuare le predizioni utilizzando il modello addestrato.

Nel metodo update_weights, vengono calcolati i gradienti utilizzando la regolarizzazione elastica. Viene calcolato il valore previsto (Y_pred) e vengono aggiornati i pesi (self.W) e il termine di bias (self.b) in base ai gradienti calcolati.

Infine, il metodo predict calcola le predizioni utilizzando self.W e self.b


Dopo l'implementazione, come prima procederò con la validazione degli iperparametri

```python
class ElasticRegression() :

    def __init__( self, learning_rate, num_iterations, l1_penality, l2_penality ) :

        self.learning_rate = learning_rate

        self.num_iterations = num_iterations

        self.l1_penality = l1_penality

        self.l2_penality = l2_penality

    # funzione di fit

    def fit( self, X, Y ) :

        self.m, self.n = X.shape  #shape restituisce dimensione n,m della matrice

        # inizializzo i pesi a un'array di zeri

        self.W = np.zeros( self.n )

        self.b = 0

        self.X = X

        self.Y = Y

        # qui gradient descend

        for i in range( self.num_iterations ) :

            self.update_weights()

        return self

    # funzione che uso per aggiornare i pesi per il gradient descend

    def update_weights( self ) :

        Y_pred = self.predict( self.X )

        # qui effettuo calcolo del gradiente

        dW = np.zeros( self.n )

        for j in range( self.n ) :

            if self.W[j] > 0 :

                dW[j] = ( - ( 2 * ( self.X[:,j] ).dot( self.Y - Y_pred ) ) +

                         self.l1_penality + 2 * self.l2_penality * self.W[j] ) / self.m

            else :

                dW[j] = ( - ( 2 * ( self.X[:,j] ).dot( self.Y - Y_pred ) )

                         - self.l1_penality + 2 * self.l2_penality * self.W[j] ) / self.m


        db = - 2 * np.sum( self.Y - Y_pred ) / self.m

        # aggiorna i pesi

        self.W = self.W - self.learning_rate * dW

        self.b = self.b - self.learning_rate * db

        return self

    # funzione ipotesi  h( x )

    def predict( self, X ) :

        return X.dot( self.W ) + self.b
```

```python
def rmse(y_true, y_pred):
    error = y_pred - y_true
    squared_error = np.power(error, 2)
    mean_squared_error = np.mean(squared_error)
    rmse = np.sqrt(mean_squared_error)
    return rmse

best_lr = 0
best_num_iterations = 0
best_l1_pen = 0
best_l2_pen = 0

i=0.0001
min_RMSE = 10000000000000000  #valore sicuramente maggiore del primo RMSE che trovero validando gli iperparametri

while (i <= 0.1):
  k = 1000

  while (k <= 10000):
    h = 0.0001

    while (h <= 0.1):
      j = 0.001

      while(j <= 0.1):
        elastic_reg = ElasticRegression(learning_rate = i, num_iterations = k, l1_penality = h, l2_penality = j)
        elastic_reg.fit(X2_train, y2_train)
        elastic_reg_predictions = elastic_reg.predict(X2_val)
        RMSE = rmse(y2_val, elastic_reg_predictions)
        print("l'RMSE della elastic net regression con lr = ",i," num_iterations = ",k,", l1_penality = ",h," e l2_penality = ",j," è: ",RMSE,)


        if (RMSE < min_RMSE):
          RMSE = min_RMSE
          best_lr = i
          best_num_iterations = k
          best_l1_pen = h
          best_l2_pen = j

          j *= 100

      h *= 100

    k += 5000

  i *=1000

print("\nQuindi la miglior combinazione di iperparametri è lr = ",best_lr," num_iterations = ",best_num_iterations,"  l1_penality = ",best_l1_pen," e l2_penality = ",best_l2_pen)
print("di seguito rieseguo la regressione con i migliori iperparametri fittando su X e y training + validation e predictando su X test:\n")
elastic_reg = ElasticRegression(learning_rate = best_lr, num_iterations = best_num_iterations, l1_penality = best_l1_pen, l2_penality = best_l2_pen)
elastic_reg.fit(X2_train_val, y2_train_val)
elastic_reg_predictions = elastic_reg.predict(X2_test)
RMSE = rmse(y2_test, elastic_reg_predictions)
print("l'RMSE della elastic net regression con lr = ",best_lr," num_iterations = ",best_num_iterations," l1_penality = ",best_l1_pen," e l2_penality = ",best_l2_pen, "è :" ,RMSE,)
```

#Classificazione con regressione lineare
utilizzo la seguente funzione di Treshold:
Threshold(z) = 1 se z>=0, Threshold(z) = 0 altrimenti

```python
import numpy as np

class LinearRegressionClassifier:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        # Inizializzazione dei parametri del modello
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0

        # Gradient descent
        for i in range(self.num_iterations):
            # Calcolo la funzione di predizione
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self._threshold(linear_model)

            # Calcolo l'errore tra la predizione e il valore reale
            error = y - y_predicted

            # Calcolo i gradienti per aggiornare i pesi e il bias
            dw = (1/num_samples) * np.dot(X.T, error)
            db = (1/num_samples) * np.sum(error)

            # Aggiorno i pesi e il bias
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        # Calcolo la funzione di predizione
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self._threshold(linear_model)
        return y_predicted

    def _threshold(self, z):
        # Funzione di threshold data
        return np.where(z >= 0, 1, 0)
```

#Classificazione con regressione logistica

```python
import numpy as np

class LogisticRegression:

    def __init__(self, lr=0.01, num_iterations=1000):
        self.lr = lr
        self.num_iterations = num_iterations
        self.w = None
        self.b = None

    def sigmoid(self, z):   #La funzione sigmoid, mappa i valori di input in un intervallo compreso tra 0 e 1. Quella di seguito è la formula matematica
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        # Aggiunta di una colonna di 1 al dataset per il termine di bias
        X = np.c_[np.ones((X.shape[0], 1)), X]  #concatena array lungo asse

        # Inizializzazione dei pesi
        self.w = np.zeros(X.shape[1])
        self.b = 0

        # Algoritmo di gradient descent
        for i in range(self.num_iterations):
            # Calcolo della funzione lineare z
            z = np.dot(X, self.w) + self.b

            # Applicazione della funzione di attivazione sigmoid
            a = self.sigmoid(z)

            # Calcolo del gradiente
            gradient_w = np.dot(X.T, (a - y)) / X.shape[0]
            gradient_b = np.mean(a - y)

            # Aggiornamento dei pesi e del bias
            self.w -= self.lr * gradient_w
            self.b -= self.lr * gradient_b

    def predict(self, X):
        # Aggiunta di una colonna di 1 al dataset per il termine di bias
        X = np.c_[np.ones((X.shape[0], 1)), X]  #np.c_ funzione numpy che concatena lungo asse specificato

        # Calcolo della funzione lineare z
        z = np.dot(X, self.w) + self.b

        # Applicazione della funzione di attivazione sigmoid
        a = self.sigmoid(z)

        # Classificazione dei risultati
        predictions = np.round(a)

        return predictions
```

#Regressione con KNN
In questa implementazione, la classe KNNRegressor rappresenta un regressore K-NN. Il costruttore prende in input un parametro k, che indica il numero di vicini da considerare per la regressione.

Il metodo fit viene utilizzato per adattare il regressore (il knn non viene propriamente fittato, utilizzo il nome fit per il metodo per convenzione) ai dati di addestramento. Vengono memorizzati i dati di addestramento (self.X_train) e i corrispondenti target di regressione (self.y_train).

Il metodo predict viene utilizzato come per gli altri modelli per effettuare le predizioni sui nuovi dati di input (X). Per ogni campione di input, viene calcolata la distanza tra il campione e i campioni di addestramento utilizzando la distanza euclidea. I k vicini più vicini sono selezionati sulla base delle distanze calcolate. I target di questi vicini sono quindi utilizzati per calcolare la media, che rappresenta il valore di regressione predetto per il campione di input.
Ho implementato un regressore di base, dove le medie non sono pesate, cioè non si da un peso differente ad un vicino in base a quale sia la distanza (in tal caso il vicino più vicino avrebbe un peso maggiore rispetto ad un vicino più lontano). Ogni vicino contribuisce in egual modo al calcolo della media, per portarmi, secondo la mia analisi, ad una stima più uniforme del valore di regressione, senza porre attenzione sull'importanza dei vicini.
Probabilmente dal punto di vista della bontà del risultato, un KNNRegressor che utilizza la media pesata potrebbe portare ad un esito che considererebbe maggiormente le caratteristiche dei dati (non per forza un'accuratezza migliore)

Le predizioni sono restituite come un array numpy nel metodo predict.

```python
class KNNRegressor:
    def __init__(self, k):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        y_pred = []
        for sample in X:
            dist = np.sqrt(np.sum((self.X_train - sample) ** 2, axis=1))
            indici_ordinati = np.argsort(dist)[:self.k]
            k_label_vicine = self.y_train[indici_ordinati]
            y_pred.append(np.mean(k_label_vicine))
        return np.array(y_pred)
```

#Implementazione
L'implementazione del KNNClassifier è simile rispetto a quella del KNNRegressor:
  - il regressore considera la media dei valori dei vicini per stimare il valore di output di un campione in input;
  - il classificatore non considera la media ma la maggioranza delle classi dei vicini più vicini (considerato anche che il classificatore non cerca un valore numerico in output come il regressore ma bensì cerca di assegnare una classe ad un valore)


Quindi la classe KNNClassifier che segue ha un costruttore che prende sempre un parametro k che rappresenta di nuovo il numero di vicini da considerare per la classificazione.

Il metodo fit viene utilizzato per "fittare" (anche se teoricamente come per quello precedente questo modello non viene addestrato) il classificatore sui dati. Prende in input sempre i vari set X e y, dove X rappresenta le features e y rappresenta le label di classe corrispondenti ai campioni in X (un vettore numpy di dimensione n).

Il metodo predict di nuovo prende in input un set di dati X e restituisce un vettore di label di classe predette per ogni campione in X. Per ogni campione in X, il metodo stavolta cerca i k vicini più vicini in base alla distanza euclidea tra il campione e i dati di addestramento. La classe di maggioranza tra questi k vicini viene utilizzata come label della classe predetta per il campione che stiamo analizzando. Il metodo restituisce un vettore numpy contenente le etichette di classe predette per ogni campione in X.

```python
class KNNClassifier:

    def __init__(self, k):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        y_pred = []
        for sample in X:
            dist = np.sqrt(np.sum((self.X_train - sample)**2, axis=1))
            indici_ordinati = np.argsort(dist)[:self.k]
            k_label_vicine = self.y_train[indici_ordinati]
            label_uniche, cont_label = np.unique(k_label_vicine, return_counts=True)
            label_maggioranza = label_uniche[np.argmax(cont_label)]
            y_pred.append(label_maggioranza)
        return np.array(y_pred)
```

#Regressione con Reti Neurali
Questa implementazione crea un'istanza di una rete neurale a 1 layer per addestrarla, ottenendo previsioni utilizzando il metodo predict.
Similmente alle implementazioni degli altri modelli, in questa implementazione di rete neurale troviamo:
* un costruttore (__init__) nel quale vengono inizializzati i parametri della rete neurale, cioè le dimensioni degli strati (input, hidden e output), il learning rate e il numero di epoche (cioè il numero di volte in cui il set passa attraverso la rete neurale durante il fit). Vengono anche inizializzati i pesi e i bias con valori casuali utilizzando np.random.randn. I pesi e i bias vengono generati casualmente per evitare che la rete neurale possa cadere in un qualche minimo indesiderato oppure un minimo non buono in fase di addestramento.

* La funzione sigmoid implementa la funzione di attivazione sigmoidale, che viene utilizzata per calcolare l'output dell'hidden layer.(stessa funzione matematica che ho implementato nella logistic regression)

* Il metodo forward esegue il passaggio in avanti dell'input attraverso la rete neurale. Calcola l'output dell'hidden layer utilizzando la funzione sigmoide e quindi calcola l'output finale senza applicare alcuna funzione di attivazione.

* Il metodo backward esegue la propagazione all'indietro (la backpropagation) per calcolare i gradienti dell'errore rispetto ai pesi e ai bias della rete neurale. Utilizza i valori intermedi calcolati durante il passaggio in avanti.

* Il metodo update_parameters aggiorna i pesi e i bias della rete neurale utilizzando il learning rate e i gradienti calcolati durante la backpropagation.

* Il metodo fit addestra la rete neurale eseguendo il passaggio in avanti e all'indietro (forward e back propagation) per un numero specificato di epoche. Calcola l'errore medio quadratico tra le previsioni della rete neurale e i valori di output desiderati.

* Il metodo predict viene utilizzato per fare previsioni sui nuovi dati dopo aver addestrato la rete neurale.




```python
import numpy as np

class NeuralNetworkR:
    def __init__(self, input_size, hidden_size, learning_rate, num_epochs):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.learning_rate = learning_rate
        self.num_epochs = num_epochs

        # Inizializzazione dei pesi e dei bias
        self.W = np.random.randn(self.input_size, self.hidden_size)
        self.b = np.zeros((1, self.hidden_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def forward(self, X):
        # Passaggio di input attraverso il layer, attivazione con sigmoide
        self.z = np.dot(X, self.W) + self.b
        self.a = self.sigmoid(self.z)

        return self.a

    def backward(self, X, y):
        # Calcolo del gradiente dell'errore rispetto all'output
        self.dz = (self.a.T - y).T
        self.dW = np.dot(X.T, self.dz)
        self.db = np.sum(self.dz, axis=0)

        # Calcolo del gradiente dell'errore rispetto all'input
        self.dX = np.dot(self.dz, self.W.T)


    def update_parameters(self):
        # Aggiornamento dei pesi e dei bias
        self.W -= self.learning_rate * self.dW
        self.b -= self.learning_rate * self.db

    def fit(self, X, y):
        for epoch in range(self.num_epochs):
            # Passaggio in avanti
            y_pred = self.forward(X)

            # Retropropagazione
            self.backward(X, y)

            # Aggiornamento dei parametri
            self.update_parameters()

    def predict(self, X):
        return self.forward(X).flatten()
```
